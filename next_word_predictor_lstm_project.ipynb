```python
faqs = """Artificial intelligence is changing the way people work, learn, and communicate every day. Many students are learning machine learning because it helps them solve real world problems using data. A machine learning model can recognize patterns, make predictions, and improve its performance over time. Deep learning is a branch of machine learning that uses neural networks with multiple layers. One of the most popular neural network architectures for sequence data is the Long Short-Term Memory network, commonly known as LSTM.
An LSTM model is designed to remember important information for long periods of time while forgetting unnecessary details. This ability makes it useful for natural language processing tasks such as text generation, next word prediction, language translation, and sentiment analysis. When training an LSTM model, the quality and quantity of text data play an important role in determining how well the model performs.
A student who wants to build a next word prediction model usually starts by collecting a text dataset. The text is cleaned by converting all words to lowercase, removing unwanted symbols, and splitting the text into tokens. These tokens are then converted into sequences of integers using a tokenizer. Each sequence represents a small portion of text that the model will use for learning patterns.
During training, the model receives a sequence of words as input and learns to predict the next word in the sequence. After many training iterations, the model gradually learns grammar, sentence structure, and common word relationships. The more diverse and larger the dataset is, the better the model becomes at predicting meaningful words.
Suppose a user types the sentence "The weather today is." A well-trained model might predict words such as sunny, pleasant, rainy, or cloudy depending on the training data. If the user continues writing, the model keeps predicting one word after another, creating complete sentences. This is how text generation systems work in many modern applications.
Many developers use Python together with TensorFlow or PyTorch to build deep learning models. TensorFlow provides easy-to-use APIs for creating sequential models, adding embedding layers, LSTM layers, dropout layers, and dense output layers. After defining the model architecture, developers compile the model using an optimizer like Adam and a suitable loss function such as categorical cross entropy.
Training a neural network requires computational resources. A computer with a dedicated GPU can significantly reduce training time compared to a CPU. However, many beginners start with cloud platforms like Google Colab or Kaggle because they provide free GPU access for educational purposes. Once the model is trained, it can be saved and later loaded for making predictions without retraining.
Evaluation is an important part of every machine learning project. Developers measure the model's accuracy on unseen data and observe whether it can generalize well. If the model performs poorly, they may increase the dataset size, adjust hyperparameters, add more LSTM units, or train for additional epochs. Regularization techniques like dropout can also help reduce overfitting.
Artificial intelligence continues to evolve rapidly, creating exciting opportunities in healthcare, finance, education, agriculture, and transportation. Students who practice building projects such as chatbots, recommendation systems, image classifiers, and text prediction models gain valuable practical experience. Learning by building real projects is one of the best ways to understand deep learning concepts. Consistent practice, experimentation, and curiosity help every developer become more skilled over time. Every successful machine learning project begins with clean data, careful experimentation, and a willingness to learn from mistakes. As technology advances, intelligent systems will continue to assist people in solving complex problems and improving everyday life.
"""
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer()
```

```python
tokenizer.fit_on_texts([faqs])
```

```python
len(tokenizer.word_index)
```

```python
input_sequences = []
for sentence in faqs.split('\n'):
  tokenized_sentence = tokenizer.texts_to_sequences([sentence])[0]

  for i in range(1,len(tokenized_sentence)):
    input_sequences.append(tokenized_sentence[:i+1])
```

```python
input_sequences
```

```python
max_len = max([len(x) for x in input_sequences])
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_input_sequences = pad_sequences(input_sequences, maxlen = max_len, padding='pre')
```

```python
padded_input_sequences
```

```python
X = padded_input_sequences[:,:-1]
```

```python
y = padded_input_sequences[:,-1]
```

```python
X.shape
```

```python
y.shape
```

```python
from tensorflow.keras.utils import to_categorical
y = to_categorical(y,num_classes=315)
```

```python
y.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
```

```python
model = Sequential()
model.add(Embedding(315, 100, input_length=100 ,input_shape=(100,))) # just put input_shape here

model.add(LSTM(150,return_sequences=True)) #  no. of  nodes  in every LSTM  gates
model.add(LSTM(150))
model.add(Dense(315, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X,y,epochs=100)
```

```python
import time
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer

# Initialize and fit the tokenizer (assuming 'faqs' is defined in a previous cell)
tokenizer = Tokenizer()
tokenizer.fit_on_texts([faqs])

text = "Artificial intelligence"

for i in range(10):
  # tokenize
  token_text = tokenizer.texts_to_sequences([text])[0]
  # padding
  padded_token_text = pad_sequences([token_text], maxlen=56, padding='pre')
  # predict
  pos = np.argmax(model.predict(padded_token_text))

  for word,index in tokenizer.word_index.items():
    if index == pos:
      text = text + " " + word
      print(text)
      time.sleep(2)
```
